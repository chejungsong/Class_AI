```python
import os, sys
import cv2

filelist = os.listdir('img/handsome_more')
cascade_file = r"C:\Users\Sub\Anaconda3\envs\sub\Lib\site-packages\cv2\data\haarcascade_frontalface_default.xml"
flag = 1

"""
filelist = []
filelist = ['angrypeople', 'ParkHyeiSoon', 'OhDalSoo','YouHaeJin', 'JungJunHa' ,'JoSaeHo', 'JoHoMin']
"""
# 파일 순서대로 파일 안에있는 이미지 이름 가져옴.
for file in filelist :
    imglist = os.listdir('img/handsome_more/'+file)
    

    
    #이미지를 순서대로 가져와서 얼굴 인식.
    for imglists in imglist :
        img_file = "img/handsome_more/"+file+"/"+imglists
        image = cv2.imread(img_file)
        image_gs = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        cascade = cv2.CascadeClassifier(cascade_file)
        face_list = cascade.detectMultiScale(image_gs,scaleFactor = 1.1, minNeighbors = 4, minSize = (70,70))
        
        overlap_naming = 1
        print(flag)
        flag += 1
        
        #인식된 얼굴을 잘라내서 저장.
        for face in face_list:
            x,y,w,h = face
            resizing_img = cv2.resize(image[y:y+h, x:x+w], dsize=(200,200), interpolation = cv2.INTER_AREA)
            cv2.imwrite("img/handsomeface/"+file+"_"+str(overlap_naming)+"_"+imglists, resizing_img)
            overlap_naming += 1

    
    print(file)
```
